```
# Let's read using timestampAsOf to count the number of rows
df = spark.read \
  .format("delta") \
  .option("timestampAsOf", "2022-06-10T02:13:28.000+0000") \
  .load("/mnt/data/data/Customer/delta/")

print(df.count())
```
